#Text Spliting
after loading the data the next step is the text spliting.
the reason for this is because LLM's have token Size limits, that is why we reduce the large text size into smaller chunks.
langchain provides a simple API for text spliter classes

```python
from langchain.text_splitter import CharacterTextSplitter

splitter = CharacterTextSplitter(
    separator='\n',
    chunk_size=200,
    chunk_overlap=0
)

chunks = splitter.split_text(str(data))
```

```python
len(chunks)
```

```python
for chunk in chunks:
  print(len(chunk))
```

#Vector database
for this project we would be using FAISS(Facebook AI similarity Search) is a library that allows quick search for embeddings for multimedia documents that are similar to each other
It solves limitations of traditional query search engines that are optimized for hash-based Searches, and provides more scalable similarity Search functions.

We would use this to convert the chunks into embeddings and we store them into FAISS index
We are using the hugging face sentence transformers

```python
# Reset the index to create a numerical 'id' column
df.reset_index(drop=True, inplace=True)

# Rename the existing column to 'text'
df.rename(columns={df.columns[0]: 'text'}, inplace=True)

# Create an 'id' column with numerical values
df['id'] = df.index + 1  # You can adjust this if you want the index to start from 0

# Reorder columns if needed
df = df[['id', 'text']]

# Display the updated DataFrame
df.head()
```

#Retrival of Questions and answers with SourceChain

Once the Vectors as stored in Vector Dataabse, the next component would be asking the questions and retriving all the relevant chunks
If the relavant chuck as in 1, and 25,,this chunks would be used to form an LLM prompt  and sent to the LLM
This reduces the Token Limit and saves cost on the Open API call.

```python
!pip install python-dotenv
!pip install -U langchain-openai
```

```python
import os
import pickle
import time
from langchain import OpenAI
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.chains.qa_with_sources.loading import load_qa_with_sources_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import OpenAIEmbeddings
from langchain.document_loaders import UnstructuredURLLoader
from langchain.vectorstores import FAISS
from dotenv import load_dotenv
```

```python
dot = "/content/drive/MyDrive/.env"
load_dotenv(dot)
openai_api_key = os.getenv("OPENAI_API_KEY")
```

```python
try:
    # Check if the OpenAI API key is available
    if not openai_api_key:
        raise ValueError("OpenAI API key is missing. Please set it in the .env file.")

    # Instantiate the OpenAI class with the API key
    llm = OpenAI(openai_api_key=openai_api_key, temperature=0.2, max_tokens=400)

    loaders = UnstructuredURLLoader(urls=[
        "https://finance.yahoo.com/news/stock-market-today-nasdaq-surges-as-chipmakers-shine-apple-pops-210312922.html?guce_referrer=aHR0cHM6Ly93d3cuZ29vZ2xlLmNvbS8&guce_referrer_sig=AQAAAKuBONPPtytsDqf50EPNfhZ1YnZyg8LLo_FpPuanAyrEzAseR5Ho2MJlcrtIiB69S43Z8KqaSNfocJA20bBFkiyik_Io_6ELWGlXLaHChboDKw_PFDwZZTrBKuZQUE7AOasMfh2TQgfOFME94utvBLvUdAZ0r1GTbEGbqmF3YjqO",
        "https://apnews.com/article/stock-market-today-china-prime-loan-rate-30dbb9cea1bd5866738d73a46fb4389d"
    ])

    print("Before loaders.load()")
    dataII = loaders.load()
    print("After loaders.load()")
    print(len(dataII))

except Exception as e:
    print(f"An exception occurred: {e}")
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=200
)

docs = text_splitter.split_documents(dataII)
len(docs)
```

```python
docs[1]
```

```python
#Create an OpenAI embeddings
embeddings = OpenAIEmbeddings()
vectorindex_openai = FAISS.from_documents(docs, embeddings)
```

```python
file_path="vector_index.pkl"
with open(file_path, 'wb') as f:
  pickle.dump(vectorindex_openai, f)
```

```python
if os.path.exists(file_path):
  with open(file_path, 'rb') as f:
    vectorIndex = pickle.load(f)
```

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorIndex.as_retriever())
chain
```

```python
query ='whay are the stock of apple rizing?'

langchain.debug=True

chain({'question': query}, return_only_outputs=True)
```
